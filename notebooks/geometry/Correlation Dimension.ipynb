## Correlation dimension

There exists a variety of abstract dimensionality measures which are compatible with our well-known intuition about dimension but also extend to more abstract spaces. One usually has to consider a tradeoff between these measures in terms of usability and generality, thus, there exist dimension measures which are easy to apply numerical but fail for some pathological spaces. In this notebook we commit ourselves to the correlation dimension measure.

Given a set of $N$ points $\boldsymbol{x}_1,\dots,\boldsymbol{x}_N$, we define the correlation sum as,
$$
C(\epsilon)
=
\frac{2}{N(N-1)}\sum^N_{i=1}\sum^N_{j=i+1}\theta\left(\epsilon-||\boldsymbol{x}_i-\boldsymbol{x}_j||\right),
$$
where $\theta$ denotes the Heaviside step function. We can read the correlation sum as a way to count the pairs $\boldsymbol{x}_i,\boldsymbol{x}_j$ where the distance between each pair is smaller than $\epsilon$.

The correlation dimension can now be obtained through,
$$
D=\lim_{\epsilon\to0}\lim_{N\to\infty}\frac{\partial \ln C(\epsilon) }{\partial \ln \epsilon},
$$ as $C(\epsilon)\varpropto\epsilon^D$ for $N\gg1$.
One practical approach to obtain the correlation dimension is to create log-log plot on $C(\epsilon)$ and fit a linear curve to find $D$.

We will now use this method to find the dimensionality of an rectangle embedded in $\mathbb{R}^2$.

```python
x1 = np.random.uniform(0, 3, 10000)
x2 = np.random.uniform(0, 1, 10000)

x = np.stack([x1, x2]).T
```

```python
data = [
    go.Scatter(x=x1, y=x2, mode='markers')
]

figure = go.Figure(data=data, layout=go.Layout(
    showlegend=False,
    xaxis=dict(title='x1'),
    yaxis=dict(title='x2', scaleanchor='x')
))

py.iplot(figure)
```

```python
epsilon = np.linspace(0, 6, 1000)

csum = np.array([correlation_sum(x1.tolist(), x2.tolist(), eps) for eps in np.nditer(epsilon)])
```

```python
data = [
    go.Scatter(x=epsilon, y=csum, name='data'),
    go.Scatter(x=epsilon, y=epsilon**2, name='fit'),
]

figure = go.Figure(data=data, layout=go.Layout(
    showlegend=True,
    xaxis=dict(title='log()', type='log'),
    yaxis=dict(title='log(C())', type='log')
))

py.iplot(figure)
```

We expect $D=2$ (orange). The data (blue) shows saturation for $\epsilon\gg1$ and $\epsilon\ll1$ which can be explained thereby that for large $\epsilon$ the balls cover already all points whereas for very small $\epsilon$ the density in between the points is decreasing.
